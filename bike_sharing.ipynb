## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana pengaruh suhu terhadap total peminjaman sepeda?

#### Membuat Clustering untuk Temperatur

````python
hour_df["temp_bins"] = pd.cut(hour_df["temp"], bins=11, labels=range(0, 11))
````

#### Membuat Grafik Hubungan Antara Temperatur dengan Peminjaman Sepeda

````python
plt.figure(figsize=(10, 6))
sns.lineplot(x='temp_bins', y='cnt', data=hour_df, marker='o')  # marker='o' untuk menunjukkan titik data
plt.title('Hubungan Antara Temperatur dengan Peminjaman Sepeda')
plt.xlabel('Temperatur (Normalized-Binned)')
plt.ylabel('Peminjaman Sepeda (cnt)')
plt.grid()  # Menambahkan grid untuk kemudahan visual
plt.show()
````

### Pertanyaan 2: Bagaimana pengaruh musim terhadap total peminjaman sepeda?

#### Menamai Setiap Musim

````python
season_mapping = {1: 'Musim Semi', 2: 'Musim Panas', 3: 'Musim Gugur', 4: 'Musim Dingin'}
day_df['season_name'] = day_df['season'].map(season_mapping)
````

````python
season_rentals = day_df.groupby('season_name')['cnt'].sum()
````

#### Membuat Grafik Hubungan Antara Musim terhadap Total Peminjaman Sepeda

````python
plt.figure(figsize=(8, 8))
plt.bar(season_rentals.index, season_rentals, color=['#4D77FF', '#FF9999', '#66FF66', '#FFCC66'])
plt.title('Distribusi Peminjaman Sepeda Lintas Musim')
plt.xlabel('Musim')
plt.ylabel('Total Peminjaman Sepeda')
plt.xticks(rotation=45)  # Putar label sumbu x jika perlu
plt.grid(axis='y')  # Menambahkan grid pada sumbu y untuk kemudahan visual
plt.show()
````

### Pertanyaan 3: Bagaimana performa peminjaman sepeda dari tahun ke tahun?

#### Menamai Setiap Tahun

````python
year_mapping = {0: '2011', 1: '2012'}
hour_df['year_name'] = hour_df['yr'].map(year_mapping)
year_rentals = hour_df.groupby('year_name')['cnt'].sum()
````

#### Membuat Grafik Hubungan Antara Tahun terhadap Total Peminjaman Sepeda

````python
plt.figure(figsize=(8,8))
plt.pie(year_rentals, labels=year_rentals.index, autopct='%1.1f%%', startangle=90, colors=['#4D77FF','#FF9999','#66FF66','#FFCC66'])
plt.title('Distribusi Peminjaman Sepeda Lintas Tahun')
plt.axis('equal')
plt.show()
````

#### Menamai Setiap Bulan

````python
# Pisahkan data berdasarkan tahun
data_2011 = hour_df[hour_df['yr'] == 0]  # 2011
data_2012 = hour_df[hour_df['yr'] == 1]  # 2012

# Agregasi peminjaman sepeda berdasarkan bulan untuk setiap tahun
rentals_2011 = data_2011.groupby('mnth')['cnt'].sum()
rentals_2012 = data_2012.groupby('mnth')['cnt'].sum()

# Labels for months
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
````

#### Membuat Grafik Hubungan Antara Tahun dan Bulan terhadap Total Peminjaman Sepeda

````python
# Membuat dua pie chart bersebelahan
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Pie chart untuk tahun 2011
ax1.pie(rentals_2011, labels=months, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
ax1.set_title('Distribusi Peminjaman Sepeda Tahun 2011')
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Pie chart untuk tahun 2012
ax2.pie(rentals_2012, labels=months, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
ax2.set_title('Distribusi Peminjaman Sepeda Tahun 2012')
ax2.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Menampilkan chart
plt.tight_layout()
plt.show()
````
